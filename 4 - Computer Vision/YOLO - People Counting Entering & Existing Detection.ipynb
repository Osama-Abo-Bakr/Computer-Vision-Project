```python

# --------------- People Counting Entering & Existing Detection -------------------
import numpy as np
from ultralytics import YOLO
import cv2
from tracker import *
import pandas as pd


video = cv2.VideoCapture(r'D:\Courses language programming\7_Computer Vision\Computer Vision - Full Course\7-YOLO ALgorithms\cctv\peoplecount1.mp4')
model = YOLO(r'yolov8s.pt')
class_list = model.names

area1 = [(312, 388), (289, 390), (474, 469), (497, 462)]
area2 = [(279, 392), (250, 397), (423, 477), (454, 469)]


entering = set()
entering_people = {}

exiting = set()
exiting_people = {}

tracker = Tracker()

while True:
    _, image = video.read()
    if image is None:
        break

    image = cv2.resize(image,(1020,500))
    h, w, _ = image.shape
    result = model.predict(image)
    a = result[0].boxes.data
    px = pd.DataFrame(a).astype(float)

    list = []
    for index, row in px.iterrows():
        x1, y1, x2, y2 = int(row[0]), int(row[1]), int(row[2]), int(row[3])
        cls = class_list[int(row[5])]
        confidence = row[4]
        if cls == 'person':
            list.append([x1, y1, x2, y2])

    people_boxes = tracker.update(list)
    for box in people_boxes:
        x11, y11, x22, y22, idd = box

        # people entering
        results = cv2.pointPolygonTest(np.array(area2, np.int32), (x22, y22), False)
        if results >= 0:
            entering_people[idd] = (x22, y22)
        if idd in entering_people:
            results1 = cv2.pointPolygonTest(np.array(area1, np.int32), (x22, y22), False)
            if results1 >= 0:
                entering.add(idd)

        # people exiting
        results2 = cv2.pointPolygonTest(np.array(area1, np.int32), (x22, y22), False)
        if results2 >= 0:
            exiting_people[idd] = (x22, y22)
        if idd in exiting_people:
            results3 = cv2.pointPolygonTest(np.array(area2, np.int32), (x22, y22), False)
            if results3 >= 0:
                exiting.add(idd)

        cv2.rectangle(image, (x11, y11), (x22, y22), (0, 165, 255), 2)


    cv2.polylines(image, [np.array(area1, np.int32)], True, (255, 0, 0), 2)
    cv2.putText(image, str('out'), (504, 471), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
    cv2.polylines(image, [np.array(area2, np.int32)], True, (255, 0, 0), 2)
    cv2.putText(image, str('In'), (466, 485), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
    cv2.putText(image, 'Number of entering people= ' + str(len(entering)),
                (20, 44), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(image, 'Number of exiting people= ' + str(len(exiting)),
                (20, 82), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 2)


    cv2.imshow('RGB', image)
    if cv2.waitKey(1) & 0xFF == ord('o'):
        break

cv2.destroyAllWindows()
video.release()
```
