```python
from ultralytics import YOLO
import cv2
import pandas as pd
import time


def countdown(image, seconds):
    h, w, _ = image.shape
    for i in range(seconds, -1, -1):
        cv2.putText(image, f'Starting in {i}...', (w//2 - 120, h//2), cv2.FONT_ITALIC, 1, (0, 0, 255), 2)
        cv2.imshow('image', image)
        cv2.waitKey(1000)  # Wait for 1000 ms (1 second)
        image[:, :, :] = 0  # Clear the text


video = cv2.VideoCapture(0)
model = YOLO(r'C:\Users\osama\Downloads\best.pt')
class_list = {0: 'Paper', 1: 'Rock', 2: 'Scissors'}

r_s = 0
l_s = 0

while True:
    _, image = video.read()
    h, w, c = image.shape

    image = cv2.flip(image, 1)
    result = model.predict(image)

    cv2.rectangle(image, (w//2-85, 0), (w//2+85, 50), (0, 165, 255), -1)
    cv2.line(image, (w//2, 0), (w//2, h), (50, 50, 50), 3)
    cv2.putText(image, str(r_s), (w//2-70, 40), cv2.FONT_ITALIC, 1.5, (150, 150, 150), 3)
    cv2.putText(image, str(l_s), (w//2+35, 40), cv2.FONT_ITALIC, 1.5, (150, 150, 150), 3)

    cv2.putText(image, 'Left', (10, 50), cv2.FONT_ITALIC, 1.5, (0, 0, 0), 3)
    cv2.putText(image, 'Right', (w-150, 50), cv2.FONT_ITALIC, 1.5, (0, 0, 0), 3)


    a = result[0].boxes.data
    px = pd.DataFrame(a).astype(float)

    score_check = []

    for index, row in px.iterrows():
        x1, y1, x2, y2 = int(row[0]), int(row[1]), int(row[2]), int(row[3])
        cls = class_list[int(row[5])]
        score_check.append((cls, x1))
        confidence = row[4]

        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 165, 255), 2)
        cv2.rectangle(image, (x1, y1 - 25), (x1 + 150, y1), (0, 165, 255), -1)

        cv2.putText(image, f'{cls}', (x1, y1 - 10), cv2.FONT_ITALIC, 0.7,
                    (0, 0, 0), 2)

        cv2.putText(image, f'{round(confidence, 2)}', (x1 + 95, y1 - 10), cv2.FONT_ITALIC, 0.7,
                    (0, 0, 0), 2)

    print(score_check)

    if len(score_check) == 2:
        if score_check[0][0] == score_check[1][0]:
            cv2.putText(image, f'Draw', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)

        if score_check[0][0] == 'Paper' and score_check[1][0] == 'Rock':
            cv2.putText(image, f'Paper Win', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)
            if score_check[0][1] >= w//2:
                l_s += 1
                countdown(image, 3)

            elif score_check[0][1] <= w//2:
                r_s += 1
                countdown(image, 3)

        if score_check[0][0] == 'Rock' and score_check[1][0] == 'Paper':
            cv2.putText(image, f'Paper Win', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)
            if score_check[1][1] <= w//2:
                r_s += 1
                countdown(image, 3)

            elif score_check[1][1] >= w//2:
                l_s += 1
                countdown(image, 3)

        if score_check[0][0] == 'Paper' and score_check[1][0] == 'Scissors':
            cv2.putText(image, f'Scissors Win', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)
            if score_check[1][1] >= w//2:
                l_s += 1
                countdown(image, 3)

            elif score_check[1][1] <= w//2:
                r_s += 1
                countdown(image, 3)

        if score_check[0][0] == 'Scissors' and score_check[1][0] == 'Paper':
            cv2.putText(image, f'Scissors Win', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)
            if score_check[0][1] <= w//2:
                r_s += 1
                countdown(image, 3)

            elif score_check[0][1] >= w//2:
                l_s += 1
                countdown(image, 3)

        if score_check[0][0] == 'Rock' and score_check[1][0] == 'Scissors':
            cv2.putText(image, f'Rock Win', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)
            if score_check[0][1] >= w//2:
                l_s += 1
                countdown(image, 3)

            elif score_check[0][1] <= w//2:
                r_s += 1
                countdown(image, 3)

        if score_check[0][0] == 'Scissors' and score_check[1][0] == 'Rock':
            cv2.putText(image, f'Rock Win', (w//2-85, h//2-30), cv2.FONT_ITALIC, 2,
                        (50, 165, 255), 3)
            if score_check[1][1] <= w//2:
                r_s += 1
                countdown(image, 3)

            elif score_check[1][1] >= w//2:
                l_s += 1
                countdown(image, 3)

    cv2.imshow('image', image)
    if cv2.waitKey(1) == ord('o'):
        break

cv2.destroyAllWindows()
video.release()
```
