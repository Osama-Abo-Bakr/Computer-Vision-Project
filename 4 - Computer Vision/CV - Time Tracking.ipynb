```python
import cv2
import time
import mediapipe as mp
from playsound import playsound
import pyttsx3
```

```python
video = cv2.VideoCapture(0)
start_time = time.time()
max_time = 10

mp_face = mp.solutions.face_detection.FaceDetection()
```

```python
while True:
    _, image = video.read()
    image = cv2.resize(image, (800, 600))
    x, y, z = image.shape


    result = mp_face.process(image)

    if result.detections:
        current_time = int((time.time() - start_time))
        if current_time > max_time:
            cv2.rectangle(image, (0, 0), (x, y), (0, 0, 255), 10)
            cv2.setWindowProperty("image", cv2.WND_PROP_TOPMOST, 1)

            cv2.putText(image, "Close Your Fucking Window.", (x//2-150, y//2), cv2.FONT_ITALIC, 1, (50, 0, 255), 2)
            input_string = "Close Your Fucking Window"
            engine = pyttsx3.init()
            engine.say(input_string)
            engine.runAndWait()

        cv2.putText(image, f"{current_time} - Second",(20, 40), cv2.FONT_ITALIC, 1, (0, 255, 0), 2)

    else:
        print(round(time.time() - start_time))
        start_time = time.time()
        # break


    cv2.imshow("image", image)

    if cv2.waitKey(1) == ord("o"):break

video.release()
cv2.destroyAllWindows()
playsound(r"D:\output.mp3")
```
