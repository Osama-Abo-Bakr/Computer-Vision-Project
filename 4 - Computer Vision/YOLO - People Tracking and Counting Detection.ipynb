```python
# --------------- People Tracking and Counting Detection -------------------
import numpy as np
from ultralytics import YOLO
import cv2
from tracker import *
import pandas as pd


def RGB(event, x, y, flags, param):
    if event == cv2.EVENT_MOUSEMOVE:
        colorsBGR = [x, y]
        print(colorsBGR)


cv2.namedWindow('RGB')
cv2.setMouseCallback('RGB', RGB)



video = cv2.VideoCapture(r'D:\Courses language programming\7_Computer Vision\Computer Vision - Full Course\7-YOLO ALgorithms\cctv\cctv.mp4')
model = YOLO(r'yolov8n.pt')
class_list = model.names



area = [(439, 400), (450, 429), (630, 300), (630, 270)]


c = set()
tracker = Tracker()

while True:
    _, image = video.read()
    if image is None:
        break
    image = cv2.resize(image, (1280, 640))
    h, w, _ = image.shape
    cv2.polylines(image,[np.array(area,np.int32)],True,(0,255,0),2)

    result = model.predict(image)
    a = result[0].boxes.data
    px = pd.DataFrame(a).astype(float)


    l = []
    for index, row in px.iterrows():
        x1, y1, x2, y2 = int(row[0]), int(row[1]), int(row[2]), int(row[3])
        cls = class_list[int(row[5])]
        confidence = row[4]
        if cls == 'person': l.append([x1, y1, x2, y2])

    people_boxes = tracker.update(l)
    for box in people_boxes:
        x11, y11, x22, y22, idd = box
        results = cv2.pointPolygonTest(np.array(area, np.int32), (x22, y22), False)
        if results >= 0:
            c.add(idd)


    cv2.rectangle(image, (x11, y11), (x22, y22), (0, 165, 255), 2)
    cv2.rectangle(image, (x11, y11 - 25), (x11 + 200, y11), (0, 165, 255), -1)
    cv2.putText(image, f'Count People: {len(c)}', (10, 25), cv2.FONT_ITALIC, 0.7, (0, 165, 255), 2)


    cv2.imshow('RGB', image)
    if cv2.waitKey(1) == ord('o'):
        break

cv2.destroyAllWindows()
video.release()
```
